```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import datetime, os

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
```

```python
application_df = application_df.drop(columns=["STATUS", "SPECIAL_CONSIDERATIONS"])
```

```python
# getting rid of extra categories in affiliation colimn
affiliation_to_replace = ["Family/Parent", "National", "Regional"]

for aff in affiliation_to_replace:
  application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(aff,"Other")

#checking the value count of the new column
application_df["AFFILIATION"].value_counts()
```

```python
# checking our USE_CASE categoriey number
application_df["USE_CASE"].value_counts()
```

```python
# getting rid of extra categories in USE_CAE colimn
use_to_replace = ["CommunityServ", "Heathcare"]

for aff in use_to_replace:
  application_df['USE_CASE'] = application_df['USE_CASE'].replace(aff,"Other")

#checking the value count of the new column
application_df["USE_CASE"].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# application_types_to_replace = application_df['APPLICATION_TYPE'].unique().tolist()
# application_types_to_replace.remove("T3")
application_types_to_replace = ["T9", "T13", "T12", "T2", "T25", "T14", "T29", "T15", "T17"]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_count = application_df['CLASSIFICATION'].value_counts()
classification_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
above_one_class = classification_count[classification_count>1]
above_one_class
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_count[classification_count<1500].index


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_dummies = pd.get_dummies(application_df, dtype=float)
df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df_dummies['IS_SUCCESSFUL'].values.reshape(-1,1)
X = df_dummies.copy()
X.drop('IS_SUCCESSFUL', axis=1, inplace=True)

# # Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 150
hidden_nodes_layer2 = 150
hidden_nodes_layer3 =30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="tanh")
)

# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="tanh")
    )

# # third hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="tanh")
)
# nn.add(
#     tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="tanh")
# )


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
